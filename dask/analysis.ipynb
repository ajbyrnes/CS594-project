```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# dask_sample_timing.csv

```python
dst_df = pd.read_csv('logs/dask_sample_timing.csv')
```

```python
dst_df.groupby('nrows')['time (s)'].mean()
```

```python
ax = sns.lineplot(data=dst_df, x='nrows', y='time (s)', marker='o')
ax.set_xscale('log')
```

```python
ax = sns.boxplot(data=dst_df, x='nrows', y='time (s)')
```
