```python
import pandas as pd
```

# dask_sample_timing.log

```python
f= open("logs/dask_sample_timing.log","r")
data = f.read()
f.close() 
```

```python
df = pd.DataFrame(columns=['nrows', 'time (s)', 'trial'])

samples = data.split('\nEND ROW\n')

for i, nrows in enumerate([1000, 10_000, 100_000, 1_000_000, 10_000_000, 100_000_000]):
    sample = samples[i] 
    s = pd.Series(sample.split('\n'))

    regex = r'Elapsed time: (\d+\.\d+) seconds'
    times = s.str.extractall(regex)[0].values

    dfi = pd.DataFrame({'nrows': nrows, 'time (s)': times, 'trial': range(len(times))})
    df = pd.concat([df, dfi], ignore_index=True)
  
df.to_csv('logs/dask_sample_timing.csv', index=False)  
df
```
